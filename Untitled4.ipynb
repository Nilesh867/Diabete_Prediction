**OBJECTIVE** : HERE WE HAVE USED SOME LIBRARIES TO FIND WHETHER ANY PATIENT HAVE DIABETES OR NOT

```python
import pandas as pd
```

```python
import numpy as nm
```

```python
diabetics = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/a6bcba4b6f9b87d8f924df1dacad300785571cfe/Diabetes.csv')
```

```python
y =diabetics['diabetes']
```

```python
y.shape
```

```python
y
```

```python
x = diabetics[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age']]
```

```python
x.shape
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state = 2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

```python
lr.predict_proba(x_test)
```
